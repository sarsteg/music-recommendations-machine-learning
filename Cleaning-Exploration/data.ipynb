```python
# Imports
import pandas as pd
import os
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Get the current directory
current_dir = os.getcwd()

#Set parent directory
parent_dir = os.path.dirname(current_dir)

# Create a path
data_file_path = os.path.join(parent_dir, 'Data', 'data.csv')

# Read the data.csv file in the Data folder and create a dataframe
data = pd.read_csv(data_file_path)
```

```python
# Show the tail of "data"
data.tail(10)
```

```python
# Show all of the data types for "data"
data.info()
```

```python
# Check for nulls
data.isnull().sum()
```

```python
# Check for duplicates
data['name'].nunique(), data.shape
```

```python
# Drop duplicate songs
data.drop_duplicates(subset=['name'], keep='first', inplace=True)
# Check for duplicates
data['name'].nunique(), data.shape
```

```python
# Drop unimportant columns from dataset "data"
data = data.drop(columns=['release_date', 'id'], axis=1)
# Reset index
data.reset_index(drop=True)
data.sort_index()
# Show data datatypes
data.info()
```

```python
# Show data
data.head()
```

```python
# Make histo grahps from data to compare data
data.hist(figsize=(15,15))

# Show the graphs
plt.tight_layout()
plt.show()
```

```python
floats = []
for col in data.columns:
  if data[col].dtype == 'float':
    floats.append(col)
len(floats)
```

```python
plt.subplots(figsize = (15, 5))
for i, col in enumerate(floats):
  plt.subplot(2, 5, i + 1)
  sb.distplot(data[col])
plt.tight_layout()
plt.show()
```

```python
# Drop columns "duration_ms", "explicit", and "instrumentalness" due to skewed data in the graphs
data = data.drop(columns=["duration_ms", "explicit", "instrumentalness"], axis=1)
data.head(10)
```

```python
# Remove "['" and "']" from data[artists]
data['artists'] = data['artists'].apply(lambda x: x.replace('[', '').replace(']',''))
data['artists'] = data['artists'].str.replace(r"^'|'$", "")
```

```python
# Show Data
data.tail(10)
```

```python
# Save new data frame 
data.to_csv('../Data/cleaned_data.csv', index=False)
```
